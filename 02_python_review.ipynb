# ## 1. Variables & Control Flow

```python
# String digunakan untuk menyimpan teks. String dapat didefinisikan dengan tanda kutip tunggal (' ') atau tanda kutip ganda (" ")
tempat_pabrik = "Sektor Dieng 1"
# Integer digunakan untuk menyimpan bilangan bulat. Integer dapat didefinisikan tanpa tanda kutip
jumlah_turbin = 7
# float digunakan untuk menyimpan bilangan desimal
kapasitas_produksi = 2.4
# Boolean digunakan untuk menyimpan nilai True atau False
status_operasi = True
# kita bisa melihat isi dan tipe data deng print() dan type()
print(tempat_pabrik, type(tempat_pabrik))
print(jumlah_turbin, type(jumlah_turbin))
print(kapasitas_produksi, type(kapasitas_produksi))
print(status_operasi, type(status_operasi))
```

```python
# kalang for digunakan untuk melakukan perulangan
for i in range(11): # range(11) digunakan untuk menghasilkan angka dari 0 sampai 10
    if i % 2 == 0: # if digunakan untuk melakukan percabangan
        print(i, "adalah bilangan genap")
    else: # else digunakan untuk melakukan percabangan jika kondisi if tidak terpenuhi
        print(i, "adalah bilangan ganjil")
```

```python
# kalang while digunakan untuk melakukan perulangan selama kondisi bernilai True
print("_" * 20)
C = -20
dC = 5
while C <= 40: # while digunakan untuk melakukan perulangan selama kondisi bernilai True
    F= (9.0/5) * C + 32 # rumus konversi celcius ke fahrenheit
    print(C, F)
    C= C + dC # increment C dengan dC
print("_" * 20)
```

## 2. functions

```python
# function digunakan untuk membuat fungsi. Fungsi dapat didefinisikan dengan kata kunci def diikuti dengan nama fungsi dan tanda kurung. Fungsi dapat memiliki parameter yang didefinisikan di dalam tanda kurung. Fungsi dapat mengembalikan nilai dengan kata kunci return
def hitung_luas_persegi(sisi): # fungsi untuk menghitung luas persegi dengan parameter sisi
    luas = sisi *sisi # menghitung luas persegi dengan rumus luas = sisi * sisi
    return luas # mengembalikan nilai luas
sisi = 5
luas_persegi = hitung_luas_persegi(sisi) # memanggil fungsi hitung_luas_persegi dengan parameter sisi
print("Luas persegi dengan sisi", sisi, "adalah", luas_persegi) # mencetak hasil luas persegi
```

## 3. Lambda & Exception Handling

```python
# lamda untuk membuat fungsi tanpa nama
# cara menggunakan lambda adalah dengan menuliskan kata kunci lambda diikuti dengan parameter dan ekspresi yang akan dikembalikan
# ekspresi di sebelah kanan tanda titik dua akan dikembalikan sebagai nilai dari fungsi lambda
kali = lambda x,y: x*y # fungsi lambda untuk mengalikan dua bilangan

# cara pemanggilan fungsi mirip seperti fungsi biasa
print(kali(5, 3)) # mencetak hasil perkalian 5 dan 3
```

```python
# Exception handling memiliki 4 blok utama yaitu try, except, else, dan finally
# try digunakan untuk mengeksekusi kode
def baca_data_sensor(data, indeks): # fungsi untuk membaca data sensor dengan parameter data dan indeks
    print(f'membaca data pada indeks {indeks}...')
    try: # try digunakan untuk mengeksekusi kode
        hasil = 100/data[indeks]
    except ZeroDivisionError: ## protokol darurat 1, menangani zeroDivisionError
        print(" -> Error: Data sensor menunjukkan angka 0. Tidak bisa dibagi 0!")
    except IndexError: ## protokol darurat 2, menangani IndexError
        print(" -> Error: Indeks yang diminta tidak ada dalam data sensor!")
    else: # else digunakan untuk mengeksekusi kode jika tidak ada error 
        print(" -> Data sensor berhasil dibaca. Hasil:", hasil)
    finally: # finally digunakan untuk mengeksekusi kode yang harus dijalankan terlepas dari ada error atau tidak
        print(" -> Proses membaca data sensor selesai.")

# Mari kita coba membaca data sensor dengan beberapa skenario
data_masuk = [8, 9, 0]
# kita panggil  fungsi dengan 3 skenario berbeda
baca_data_sensor(data_masuk, 0) # skenario normal
baca_data_sensor(data_masuk, 2) # skenario zeroDivisionError
baca_data_sensor(data_masuk, 3) # skenario IndexError


```

## 4. basic I/O

```python
# basic I/O memiliki 3 fungsi utama yaitu input(), print(), dan open()
import os
# 1. output: menampilkan data ke layar
print("="*40)
print("Sistem Pencatatan Suhu Geothermal")
print("="*40)

# 2. input: meminta data dari user
nama_operator = input("Masukkan nama operator: ")
lokasi_well = input('masukkan kode well(contoh:Sektor_A):')
suhu_teks = input('masukkan suhu dalam celcius (contoh: 25.5): ')

# mengubah data suhu dari string ke float
suhu_celcius = float(suhu_teks)

# proses logika pengolahan data suhu
status = ""
if suhu_celcius > 300.0:
    status ="KRITIS!-suhu melebihi batas aman!"
elif suhu_celcius < 150.0:
    status = "RENDAH-produksi uap tidak optimal!"
else:
    status = "NORMAL"

# 3. output: menampilkan hasil evaluasi suhu ke layar
print(f"\n[HASIL EVALUASI]: well {lokasi_well} berstatus {status}")

# 4. file I/O: menyimpan data ke file dengan 'with' statement
nama_file= "geothermal_log.txt"
with open(nama_file, 'a') as file_log: # 'a' untuk menambahkan data ke file tanpa menghapus data lama(mirip append)
    text_log = f"Operator: {nama_operator}| Well: {lokasi_well}| Suhu: {suhu_celcius}C| Status: {status}\n" # \n untuk membuat baris baru
    file_log.write(text_log) # menulis data ke file
print(f"\nsistem: data telah berhasil disimpan ke file {os.path.abspath(nama_file)}")
```

### 3 For this exercise
#### Write a function (guessing_game) that takes no arguments.
#### When run, the function chooses a random integer between 0 and 100 (inclusive).
#### Then ask the user to guess what number has been chosen.
#### Each time the user enters a guess, the program indicates one of the following:– Too high– Too low– Just right
#### If the user guesses correctly, the program exits. Otherwise, the user is asked to try again.
#### The program only exits after the user guesses correctly.

```python
import random # 
def permainan_tebak_angka():
    angka = random.randint(0, 100)
    while True:
        tebakan = int(input("masukkan angka yang anda pikirkan dari 0-100:"))
        if tebakan == angka:
            print(f"tebakan anda benar! jawaban adalah {angka}")
            break
        elif tebakan < angka:
            print(f"tebakan anda kekecilan")
        else:
            print(f"tebakan anda kebesaran!")

print(f"mari kita bermain tebak angka 0-100 yang ada dipikiran saya")
permainan_tebak_angka()
```

#### If the word begins with a vowel (a, e, i, o, or u), add “way” to the end of the word. So “air” becomes “airway” and “eat” becomes “eatway.”
#### If the word begins with any other letter, then we take the first letter, put it on the end of the word, and then add “ay.” Thus, “python” becomes “ythonpay” and “computer” becomes “omputercay.”

```python
def latin_pig():
    kata = str(input("masukkan kata yang ingin anda ubah ke kode:"))
    if kata[0] in "aiueo":
        return f"{kata}way"
    else:
        return f"{kata[1:]}{kata[0]}ay"
print(latin_pig())
```
