### List
#### wWhat is List?
List adalah struktur data yang berisi urutan nilai (sekuens) dan bersifat mutable, yang artinya elemen di dalamnya dapat diubah, ditambah, atau dihapus setelah didefinisikan

```python
tup = ('hai', 'antek', 'antek', 'asing!')
a_list = list(tup) #list dan tuple hampir mirip secara semantik (walaupun tuple immutable) dan bisa digunakan secara bergantian di bayak fungsi
a_list
```

```python
gen = range(11)
print (gen)
list (gen) # fungsi list sering digunakan dalam pemprosesan data sebagai cara untuk mematerialkan ekspresi iterator atau generator
```

```python
# karena list mutable, maka kita bisa adding and removing elements dari list
a_list.append('prabowo') # element bisa ditambahkan(append) ke list dengan method append()
print(f'list setelah append: {a_list}')
a_list.insert(4, 'kata') # dengan insert, kita bisa memasukan(insert) element ke posisi tertentu dalam list
print(f'list setelah insert: {a_list}')
a_list.pop(4) # pop adalah operasi kebalikan dari insert, yaitu menghapus element pada posisi tertentu dalam list
print(f'list setelah pop: {a_list}')
a_list.remove('antek') # remove akan mencari element pertama yang cocok dengan nilai yang diberikan dan menghapusnya dari list
print(f'list setelah remove: {a_list}')
```

```python
'hai' in a_list # operator in digunakan untuk mengecek apakah suatu nilai ada dalam list
print(f'apakah "hai" ada dalam list? {'hai' in a_list}')
'hai' not in a_list # operator not in digunakan untuk mengecek apakah suatu nilai tidak ada dalam list
print(f'apakah "hai" tidak ada dalam list? {'hai' not in a_list}')
```

```python
[10, None, 'king'] + [3, 7, 1, (67, 69)] # list dapat dikonkatenasi dengan tanda +
```

```python
x = [10, None, 'king']
x.extend([3, 7, 1, (67, 69)]) # list dapat diperluas dengan method extend()
print(x) # menggunakan extend() lebih efisien daripada menggunakan operator + karena extend() memodifikasi list yang ada, sedangkan operator + membuat list baru
```

```python
a = [3, 8, 1, 5, 2,9, 0, 7, 4, 6]
a.sort() # sort method digunakan untuk mengurutkan elemen dalam list, dan akan mengubah list yang ada
print(a)
```

```python
b = ['saw', 'small', 'he', 'foxes', 'six']
b.sort(key=len) # sort method dapat menerima argumen key untuk menentukan kriteria pengurutan, dalam hal ini berdasarkan panjang string
print(b)
```

```python
seq = [7, 2, 3, 7, 5, 6, 0, 1]
print(seq[1:5]) # dengan menggunakan :, kita bisa slicing list
seq[3:5] = [10, 20] # slicing list dapat digunakan untuk mengubah nilai dalam list, tetapi jumlah elemen yang diubah harus sama dengan jumlah elemen yang diganti
print(seq)
seq[-4:] # dengan index negatif, slicing dihitung dari akhir list
print(seq[-4:]) # menampilkan 4 elemen terakhir dari list
```

### Dictionary
#### Dictionary adalah struktur data koleksi yang bersifat mutable (bisa diubah) yang memetakan kumpulan keys (kunci) ke sekumpulan values (nilai)
kita bisa anggap key (immutabble) sebagai label loker dan values(bebas) sebgai isi loker yang bisa apa saja
penulisannya nya adalah key:values

```python
d1 = {"a" : "Hidup Jokowi", "b" : "17, 8, 1945"} # dictionary didefinisikan dengan {}
d1[3] = "bilangan bulat" # insert dan set element masih menggunakan syntax u=yang sama dengan list dan tuple
d1
```

```python
d1[5] = "suatu nilai"
print (d1)
d1["dummy"] = "nilai lain"
print (d1)
del d1[5] # del statement akan menghapus element dari dictionary tanpa mengembalikan nilai
print (d1)
print(d1.pop("dummy")) # pop akan menghapus element dari dictionary dan mengembalikan nilai yang dihapus
print(d1)
```

```python
print(list(d1.keys())) # keys() akan menampilkan semua key dalam dictionary
print(list(d1.values())) # values() akan menampilkan semua value dalam dictionary
print(list(d1.items())) # items() akan menampilkan semua pasangan key-value dalam dictionary dalam bentuk list of tuples
d1.update({"c": "nilai baru", "d": "nilai lain"}) # update() akan menambahkan atau memperbarui element dalam dictionary
d1
```

```python
# seringkali kita bertemu dengan 2 sequence yang ingin kita pasangkan menjadi dictionary
# pada percobaan pertama, mungkin kita akan menulis kode berikut:
# Mendefinisikan sequence terlebih dahulu
key_list = ['nama', 'pekerjaan', 'kota']
value_list = ['Budi', 'Developer', 'Jakarta']
mapping = {} # membuat dictionary kosong
for key, value in zip(key_list, value_list): # mengiterasi pasangan key-value dari dua list
    mapping[key] = value # menambahkan pasangan key-value ke dalam dictionary
mapping
```

```python
# tetapi tidak efisien
# karena dict pada dasarnya adalah kumpulan 2 tuple
# maka kita bisa menggunakan fungsi built-in dict() untuk membuat dictionary dari dua sequence:
tuples = zip(range(5), reversed(range(5)))
print (tuples) # akan menampilkan objek zip, yang merupakan iterator yang menghasilkan pasangan tuple dari dua sequence
mapping = dict(tuples) # fungsi dict() akan mengubah iterator zip menjadi dictionary
mapping
```

```python
# umum sekali memiliki logic seperti ini untuk membuat jawaban cadangan pada sesuatu yang belum diberi label:
some_dict = {'nama': 'Budi', 'pekerjaan': 'Developer', 'kota': 'Jakarta'}
key = 'hobi'
if key in some_dict: # memeriksa apakah key ada dalam dictionary
    value = some_dict[key] # mengambil nilai dari dictionary jika key ada
else: # jika key tidak ada dalam dictionary
    value = 'belum ada data' # mengatur nilai default
print(f"Nilai untuk key dengan while '{key}': {value}")
# tetapi kita bisa menyederhanakan dengan metode get() dan kode menjadi simpel seperti ini:
value = some_dict.get(key, 'belum ada data') # mengambil nilai dari dictionary jika key ada, jika tidak ada maka mengatur nilai default(non-None)
print(f"Nilai untuk key dengan get '{key}': {value}")
```

```python
# bagaimana jika kita mengecek key sekaligus value? maka logic dasarnya adalah seperti ini:
words = ["apple", "banana", "cherry", "date", "elderberry"] # mendefinisikan list of words
by_letter = {} # membuat dictionary kosong
for word in words: # mengiterasi setiap kata dalam list
    letter = word[0] # mengambil huruf pertama dari kata
    if letter not in by_letter: # memeriksa apakah huruf pertama sudah ada sebagai key dalam dictionary
        by_letter[letter] = [word] # jika belum ada, buat key baru dengan value berupa list yang berisi kata tersebut
    else: # jika sudah ada
        by_letter[letter].append(word) # tambahkan kata ke list yang sudah ada untuk key tersebut
print(f"Dictionary by_letter dengan kalang: {by_letter}")
# tetapi kita bisa menyederhanakan dengan metode setdefault() dan kode menjadi simpel seperti ini:
by_letter = {} # membuat dictionary kosong
for word in words: # mengiterasi setiap kata dalam list
    letter = word[0] # mengambil huruf pertama dari kata
    by_letter.setdefault(letter, []).append(word)
 # setdefault() akan mengembalikan value untuk key jika ada, jika tidak ada maka akan membuat key baru dengan value default yang diberikan (dalam hal ini list kosong), kemudian append kata ke list tersebut
print(f"Dictionary by_letter dengan setdefault: {by_letter}")
```

```python
# ada cara yang lebih efisien dan sering digunakan di data analisis
# cara tersebut membuat objek dengan nilai default secara otomatis setiap ada kata kunci baru
# cara tersebut adalah menggunakan defaultdict dari modul collections
from collections import defaultdict # pertama-tama kita import dulu
words = ["apple", "banana", "cherry", "date", "elderberry"]
by_letter = defaultdict(list) # disini kita definisikan tipe data(misal: list, int, atau float)
for word in words:
    by_letter[word[0]].append(word)
by_letter
```

### Set
sama seperti konsep himpunan di matematika, set adalah kumpulan elemen unik tidak berurutan


```python
# Set bisa dibuat dengan menggunakan kurung kurawal {} atau dengan fungsi built-in set()
set1 = {1, 2, 3, 4, 5} # set didefinisikan dengan {}
set2 = set([1, 2, 3, 4, 5]) # set juga bisa dibuat dengan fungsi built-in set()
print(f'set1: {set1}, set2: {set2}')
```

```python
# set python juga support operasi matematika set
a = {1, 2, 3, 4, 5}
b = {3, 4, 5, 6, 7}
print(f'a union b: {a | b}') # union dari dua set, operator | digunakan untuk menggabungkan dua set
print(f'a intersection b: {a & b}') # intersection dari dua set, operator & digunakan untuk mencari irisan dari dua set
```

##### Lebih lengkapnya ada pada tabel berikut:


```python
# semua operasi set memiliki lawannya
# ini memungkinkan kita mengganti konten set di sebelah kiri operasi dengan hasil operasi
c = a.copy() # membuat salinan dari set a
c |= b # union dari set c dan b, hasilnya akan menggantikan isi set c
print(f'c setelah union dengan b: {c}')
d = a.copy() # membuat salinan dari set a
d &= b # intersection dari set d dan b, hasilnya akan menggantikan isi set d
print(f'd setelah intersection dengan b: {d}')
```

```python
# seperti dictionary. elemen set biasanya harus immutable dan wajib hashable
my_data = [1, 2, 3, 4, 5] # list dengan elemen yang tidak unik
my_set = set(my_data) # mengubah list menjadi set, menghilangkan elemen duplikat
my_set
```

### Buit-in sequence functions

```python
# ada enumerate(), yang memberi nomor urut pada setiap elemen dalam iterable
collection = ['a', 'b', 'c', 'd']
# logika dasarnya seperti ini:
print("dengan logika dasar:")
index = 0
for value in collection:
    print(f'index: {index}, value: {value}')
    index += 1
print("dengan logika dasar:", index)
# kalau kita ingin menggunakan enumerate(), kita bisa menulis kode seperti ini:
print("dengan enumerate():")
for index, value in enumerate(collection):
    print(f'index: {index}, value: {value}')
```

```python
# ada sorted(), yang memberi list urut baru dari element sequence
sorted_list = sorted([7, 1, 2, 6, 0, 3, 5, 4])
print(sorted_list)
sorted_string = sorted("prabowo") # sorted() juga bisa digunakan untuk mengurutkan string
print(sorted_string)
```

```python
# ada zip(), yang menggabungkan beberapa iterable menjadi satu iterable yang menghasilkan tuple
seq1 = ['a', 'b', 'c']
seq2 = [1, 2, 3]
zipped = zip(seq1, seq2) # zip() akan menggabungkan dua iterable menjadi satu iterable yang menghasilkan tuple
print(list(zipped)) # akan menampilkan list dari pasangan tuple yang dihasilkan oleh zip()
seq3 = [True, False] # zip() bisa mengambil jumlah sembarang dari iterable, dan akan berhenti pada iterable yang paling pendek
print(list(zip(seq1, seq2, seq3))) # akan menampilkan list dari pasangan tuple yang dihasilkan oleh zip() dengan tiga iterable
#penggunaan zip() yang paling sering adalah mengiterasi beberapa sequence secara bersamaan, biasanya juga digabungkan dengan enumerate() untuk mendapatkan index dari setiap pasangan tuple yang dihasilkan oleh zip()
for index, (a, b) in enumerate(zip(seq1, seq2)):
    print(f'index: {index}, a: {a}, b: {b}')
```
